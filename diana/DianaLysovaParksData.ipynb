We are going to be using this data set: https://data.wprdc.org/dataset/parks .
I downloaded the data set and named it parks.csv . Let's load it into Jupyter.

```python
import geopandas
%matplotlib inline
import pandas as pd
from shapely.geometry import Point
df= pd.read_csv('./parks.csv')
df.head(3)
```

I also went ahead and printed the first three rows, just to make everything was working properly.

```python
print (df["neighborhood"])
```

Next, I extracted the "neighborhood" column of the data frame. This showed me all of the neighborhoods and how many parks there were overall.

```python
df1= df["neighborhood"].value_counts()
print(df1)
```

Then, finally, I found how many times each neighborhood showed up in the data set- i.e how many parks each neighborhood had. I didn't need every single value, which is why only the top 5 highest and top 5 lowest are shown.

```python
df.neighborhood.value_counts().sort_values().plot.bar(figsize=(55,50))
```

To visualize this data, I created a bar plot.

```python
df.neighborhood.value_counts().sort_values().plot.pie()
```

I also created a pie chart. This may seem convoluted since so many of the neighborhoods are condensed and you cannot see the names clearly. However, the point of this chart was to show just how many more parks Easy Liberty has. Where the other neighborhoods are small slivers, East Liberty has a much more significant part of the pie chart.
